```python
# Bloque 1

import sys
import os

# Añadir raíz del repo (peltierlab/)
sys.path.append(os.path.abspath(".."))

# Import limpio tipo paquete
from peltierlab.core.simulator_hysteresis_real import SimulatorHysteresisReal

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Parámetros 3-nodos
best_params = [1.9507, 2.4906, 36.4772, 0.4806, 14.0687, 2.2298, 66.5757, 11.8439]

T_start = 19.0
```

```python
# Bloque 2

def run_simulation_hysteresis(T_set, dT1, dT2):
    
    t_new = np.linspace(0, 600, 1200)

    sim = SimulatorHysteresisReal(best_params, T_start=T_start)

    Tc, Tm, Th, pwm = sim.simulate(
        t_custom=t_new,
        T_set=T_set,
        dT1=dT1,
        dT2=dT2,
        P_max=5.0   # 🔥 IMPORTANTE → subir potencia
    )

    plt.close('all')
    plt.figure(figsize=(10,5))

    plt.plot(t_new, Tc, linewidth=2, label='Simulated Temp')
    plt.axhline(T_set, linestyle='--', linewidth=1, label=f'Setpoint {T_set}°C')

    plt.xlabel("Tiempo [s]")
    plt.ylabel("Temperatura [°C]")
    plt.title("Hysteresis Control")
    plt.grid(True)
    plt.legend()

    plt.show()
```

```python
# Bloque 3

interact(
    run_simulation_hysteresis,
    T_set=FloatSlider(value=12.0, min=10.0, max=20.0, step=0.1),
    dT1=FloatSlider(value=0.5, min=0.1, max=2.0, step=0.1),
    dT2=FloatSlider(value=0.5, min=0.1, max=2.0, step=0.1)
)
```
