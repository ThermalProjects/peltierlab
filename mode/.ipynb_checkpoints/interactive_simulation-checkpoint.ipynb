```python
import sys, os
sys.path.append(os.path.abspath(".."))  # Para que Python encuentre peltierlab

from peltierlab.core.simulator import Simulator
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
```

```python
def run_simulation(T_set=12.0, bias=0.4):
    best_params = [1.9507, 2.4906, 36.4772, 0.4806, 14.0687, 2.2298, 66.5757, 11.8439]
    T_start = 19.0
    t_new = np.linspace(0, 300, 500)

    sim = Simulator(best_params, T_start=T_start)
    Tc_sim, pwm_sim = sim.simulate_3nodes_FOPID(t_new, T_set=T_set - bias)

    # Graficar temperatura
    plt.figure(figsize=(10,5))
    plt.plot(t_new, Tc_sim, label='Temperatura Simulada', linewidth=2)
    plt.axhline(T_set, color='red', linestyle='--', label='Setpoint')
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Temperatura [°C]")
    plt.title(f"Simulación 3-nodos + FOPID (Setpoint={T_set}°C, Bias={bias}°C)")
    plt.grid(True)
    plt.legend()
    plt.show()

    # Graficar PWM
    plt.figure(figsize=(10,3))
    plt.plot(t_new, pwm_sim, label='PWM', color='orange', linewidth=2)
    plt.xlabel("Tiempo [s]")
    plt.ylabel("PWM")
    plt.title("Señal de control PWM")
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
interact(run_simulation,
         T_set=FloatSlider(value=12.0, min=5.0, max=25.0, step=0.5, description='Setpoint [°C]'),
         bias=FloatSlider(value=0.4, min=0.0, max=2.0, step=0.1, description='Bias [°C]'));
```
