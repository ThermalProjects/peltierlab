```python
# fopid.ipynb - Bloque 1

from peltierlab.core.simulator import Simulator
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# =================== Parámetros 3-nodos reales ===================
best_params = [1.9507, 2.4906, 36.4772, 0.4806, 14.0687, 2.2298, 66.5757, 11.8439]

# Temperatura inicial
T_start = 19.0

# Configuración PID inicial
Kp_default = 58.93
Ki_default = 3.91
Kd_default = 2.66

# Valores por defecto FOPID
lambda_default = 0.67
mu_default     = 1.47
```

```python
# Bloque 2

def run_simulation_fopid(T_set, bias, Kp, Ki, Kd, lam, mu):
    t_new = np.linspace(0, 300, 500)  # tiempo [s]

    sim = Simulator(best_params, T_start=T_start)

    Tc_sim, _ = sim.simulate_3nodes_FOPID(
        t_custom=t_new,
        T_set=T_set,
        Kp=Kp, Ki=Ki, Kd=Kd,
        bias=bias,
        lam=lam, mu=mu
    )

    # ====== Solo temperatura, figura única ======
    plt.close('all')  # cierra cualquier figura anterior
    plt.figure(figsize=(10,5))
    plt.plot(t_new, Tc_sim, color='green', linestyle='-.', linewidth=2, label='Temperatura Simulada')
    plt.axhline(T_set, color='red', linestyle='--', linewidth=1, label=f'Setpoint {T_set}°C')
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Temperatura [°C]")
    plt.title("Simulación 3-nodos + FOPID")
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
# Bloque 3

interact(run_simulation_fopid,
         T_set=FloatSlider(value=12.0, min=5.0, max=25.0, step=0.5, description='Setpoint [°C]'),
         bias=FloatSlider(value=0.0, min=-2.0, max=2.0, step=0.1, description='Bias [°C]'),
         Kp=FloatSlider(value=Kp_default, min=0, max=200, step=1, description='Kp'),
         Ki=FloatSlider(value=Ki_default, min=0, max=50, step=0.1, description='Ki'),
         Kd=FloatSlider(value=Kd_default, min=0, max=50, step=0.1, description='Kd'),
         lam=FloatSlider(value=0.67, min=0.1, max=2.0, step=0.01, description='Lambda'),
         mu=FloatSlider(value=1.47, min=0.1, max=2.0, step=0.01, description='Mu'));
```
